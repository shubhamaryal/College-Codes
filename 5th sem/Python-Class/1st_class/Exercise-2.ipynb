# Exercise 2 - Grade Processor (Medium)

Create a program that:
1. has a list masrks = [78, 85, 92, 66, 88, 95, 71, 83]
2. calculates and prints:
   - average mark
   - highest and lowest mark
   - how many students got >= 80 (distinction)
3. Create a new list grades where:
   .>= 90 -> "A+" , >= 80 -> "A", >= 70 -> "B", else -> "C"
4. saves the output in a file grade_report.txt exactly like this:

```python
marks = [78, 85, 92, 66, 88, 95, 71, 83]

average = sum(marks) / len(marks)
highest = max(marks)
lowest = min(marks)

count_distinction = 0
for m in marks:
    if m >= 80:
        count_distinction += 1

with open("grade_report.txt", "w") as f:
    f.write(f"Average Marks: {average}")
    f.write("\n")
    f.write(f"Highest Marks: {highest}")
    f.write("\n")
    f.write(f"Lowest Marks: {lowest}")
    f.write("\n")
    f.write(f"Students with distinction: {count_distinction}")
    f.write("\n")
    f.write("\n")
    f.write("List of gardes: ")

    for m in marks:
        if m >= 90:
            f.write("A+")
            f.write("\n")
        elif m >= 80:
            f.write("A")
            f.write("\n")
        elif m >= 70:
            f.write("B")
            f.write("\n")
        else:
            f.write("C")
            f.write("\n")

print("grade_report.txt saved.")
```
