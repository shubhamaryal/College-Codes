# Exercise 3 – Student Dictionary Database (Medium-Hard)
Create a dictionary for one student with these keys:
name, roll_no, faculty, semester, marks (marks = list of 5 subjects)

Then write code that:
1. Inputs all data from keyboard
2. Calculates percentage
3. Adds a new key "result": "Pass" if percentage ≥ 40, otherwise "Fail"
4. Prints everything neatly
5. Appends this student record (as one line) to a file students.txt
Example line: Ram Prasad, 101, BCE, 5, [75,80,88,92,66], 80.2%, Pass

```python
name = input("Enter student name: ")
roll_no = input("Enter roll number: ")
faculty = input("Enter faculty: ")
semester = input("Enter semester: ")

marks = []
print("Enter marks for 5 subjects:")
for i in range(5):
    m = float(input(f"Subject {i+1}: "))
    marks.append(m)

student = {
    "name": name,
    "roll_no": roll_no,
    "faculty": faculty,
    "semester": semester,
    "marks": marks
}

total = sum(marks)
percentage = total / 5

if percentage >= 40:
    student["result"] = "Pass"
else:
    student["result"] = "Fail"

print("\nStudent Record")
print("---------------")
print("Name      :", student["name"])
print("Roll No   :", student["roll_no"])
print("Faculty   :", student["faculty"])
print("Semester  :", student["semester"])
print("Marks     :", student["marks"])
print(f"Percentage: {percentage:.2f}%")
print("Result    :", student["result"])

line = f"{student['name']}, {student['roll_no']}, {student['faculty']}, {student['semester']}, {student['marks']}, {percentage:.2f}%, {student['result']}\n"

with open("students.txt", "a") as f:
    f.write(line)
```
