```python
import requests
response = requests.get('https://pokeapi.co/api/v2/pokemon/pikachu')
data = response.json()
print(f"Name : {data['name']}")
print(f"Types: {[t['type']['name'] for t in data ['types']]}")
```

```python
  BASE_URL = "https://jsonplaceholder.typicode.com/posts"

# 1. GET: Retrieve data (Read)
# Fetches the post with ID 1
get_response = requests.get(f"{BASE_URL}/1")
print("--- GET Response ---")
print(get_response.json()) 
```

```python
new_post = {
    "title": "Learning REST APIs",
    "body": "This is a demonstration using Python Requests.",
    "userId": 1
}
post_response = requests.post(BASE_URL, json=new_post)
print("\n--- POST Response (Created) ---")
print(post_response.status_code) # 201 means "Created"
print(post_response.json())
```

```python
import requests

response = requests.get('https://open.er-api.com/v6/latest/USD')
data = response.json()

print(f"Base: {data['base_code']}")
print(f"1 USD = {data['rates']['NPR']:.2f} NPR")
print(f"1 USD = {data['rates']['EUR']:.2f} EUR")

# Full list (first 5)
rates = dict(list(data['rates'].items())[:5])
print("Sample rates:", rates)
```
