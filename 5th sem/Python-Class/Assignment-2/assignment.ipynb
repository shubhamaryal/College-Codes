```python
# Data processing library
import pandas as pd

# Mapping and visualization library
import folium
```

```python
# Load GPS trace dataset from CSV file
# Replace filename with your actual dataset path
df = pd.read_csv("madrid_gps_traces.csv")

# Preview first few rows to verify structure
df.head()
```

```python
# Check column names and data types
df.info()

# Check missing values count per column
df.isnull().sum()
```

```python
# Remove rows with missing latitude or longitude
df = df.dropna(subset=["latitude", "longitude"])

# Remove duplicate GPS records if present
df = df.drop_duplicates()

# Reset index after cleaning
df = df.reset_index(drop=True)
```

```python
# Convert timestamp column to datetime format
df["timestamp"] = pd.to_datetime(df["timestamp"])

# Sort data by time to maintain correct movement order
df = df.sort_values("timestamp")
```

```python
# Extract hour from timestamp for time-based grouping
df["hour"] = df["timestamp"].dt.hour

# Extract date separately if needed
df["date"] = df["timestamp"].dt.date
```

```python
# Approximate Madrid bounding box
lat_min, lat_max = 40.2, 40.6
lon_min, lon_max = -3.9, -3.5

# Filter rows within bounding box
df = df[
    (df["latitude"] >= lat_min) &
    (df["latitude"] <= lat_max) &
    (df["longitude"] >= lon_min) &
    (df["longitude"] <= lon_max)
]
```

```python
# Madrid city center coordinates
madrid_center = [40.4168, -3.7038]

# Create folium map centered at Madrid
m = folium.Map(location=madrid_center, zoom_start=12)
```

```python
# Add GPS points to map
for _, row in df.iterrows():
    folium.CircleMarker(
        location=[row["latitude"], row["longitude"]],
        radius=2,                 # small radius for dense data
        color="blue",
        fill=True,
        fill_opacity=0.6
    ).add_to(m)
```

```python
# Create coordinate list for path drawing
path_coords = df[["latitude", "longitude"]].values.tolist()

# Draw path line across GPS points
folium.PolyLine(
    path_coords,
    color="red",
    weight=3,
    opacity=0.8
).add_to(m)
```
