```python
from collections import Counter, defaultdict

# Frequency of categories
tags = ['python', 'data', 'python', 'ml', 'data', 'python']
tag_counts = Counter(tags)
# Result: Counter({'python': 3, 'data': 2, 'ml': 1})

# Grouping indices by value
data_map = defaultdict(list)
for index, value in enumerate(tags):
    data_map[value].append(index)

print(tag_counts)
# Output: Counter({'python': 3, 'data': 2, 'ml': 1})

print(data_map)
# Output: defaultdict(<class 'list'>, {'python': [0, 2, 5], 'data': [1, 4], 'ml': [3]})
```
